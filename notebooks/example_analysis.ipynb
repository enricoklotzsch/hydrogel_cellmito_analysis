# Workspace Setup and Validation

This notebook demonstrates how to automate and validate the setup of a new workspace for image analysis of cellular morphology and mitochondria networks in TIFF stacks.

## 1. Install and Import Required Libraries

Install and import the necessary Python packages for workspace setup and image analysis.

```python
# Install required packages (uncomment if running for the first time)
# !pip install numpy scipy scikit-image matplotlib tifffile

import os
import sys
import numpy as np
import tifffile
import matplotlib.pyplot as plt
from skimage import filters, measure, morphology
from pathlib import Path
```

## 2. Initialize a New Workspace Directory

Create a new directory for the workspace if it does not already exist.

```python
workspace_dir = Path('Image_Analysis_Raphael')
if not workspace_dir.exists():
    workspace_dir.mkdir()
    print(f"Created workspace directory: {workspace_dir}")
else:
    print(f"Workspace directory already exists: {workspace_dir}")
```

## 3. Create and Configure Virtual Environment

Use the `venv` module to create a virtual environment and install essential packages.

```python
import venv
venv_dir = workspace_dir / 'venv'
if not venv_dir.exists():
    venv.create(venv_dir, with_pip=True)
    print(f"Created virtual environment at {venv_dir}")
else:
    print(f"Virtual environment already exists at {venv_dir}")
```

## 4. Generate Workspace Configuration Files

Create basic configuration files such as `README.md`, `.gitignore`, and `requirements.txt` using Python file operations.

```python
# Create README.md
readme_path = workspace_dir / 'README.md'
if not readme_path.exists():
    readme_path.write_text("""# Image Analysis of Cellular Morphology and Mitochondria Networks\n\nWorkspace for analyzing cellular morphology and mitochondria networks in TIFF stacks.\n""")
    print(f"Created {readme_path}")
else:
    print(f"{readme_path} already exists")

# Create .gitignore
gitignore_path = workspace_dir / '.gitignore'
if not gitignore_path.exists():
    gitignore_path.write_text("venv/\n__pycache__/\n*.pyc\n.ipynb_checkpoints/\n")
    print(f"Created {gitignore_path}")
else:
    print(f"{gitignore_path} already exists")

# Create requirements.txt
requirements_path = workspace_dir / 'requirements.txt'
if not requirements_path.exists():
    requirements_path.write_text("numpy\nscipy\nscikit-image\nmatplotlib\ntifffile\n")
    print(f"Created {requirements_path}")
else:
    print(f"{requirements_path} already exists")
```

## 5. Verify Workspace Setup

Check that all expected files and directories exist and print a summary of the workspace structure.

```python
# List workspace contents
for path in workspace_dir.iterdir():
    print(path)

# Check for expected files and directories
expected = ['README.md', '.gitignore', 'requirements.txt', 'venv']
missing = [item for item in expected if not (workspace_dir / item).exists()]
if not missing:
    print("All expected files and directories are present.")
else:
    print(f"Missing: {missing}")
```
